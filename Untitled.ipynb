```python
from PIL import Image
from pytesseract import pytesseract

#Define path to tessaract.exe
path_to_tesseract = r'/opt/homebrew/bin/tesseract'

#Define path to image
path_to_image = r'Images/testimage1.jpeg'

#Point tessaract_cmd to tessaract.exe
pytesseract.tesseract_cmd = path_to_tesseract

#Open image with PIL
img = Image.open(path_to_image)

#Extract text from image
text = pytesseract.image_to_string(img)

print(text[:-1])
```

```python
import easyocr
reader = easyocr.Reader(['en'])
result = reader.readtext('Images/testimage1.jpeg', detail = 0)
print(result)
```

```python
import cv2

path_to_tesseract = r'/opt/homebrew/bin/tesseract'
pytesseract.tesseract_cmd = path_to_tesseract

img = cv2.imread("Images/testimage1.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU |
                                          cv2.THRESH_BINARY_INV)
cv2.imwrite('threshold_image.jpeg',thresh1)
```
